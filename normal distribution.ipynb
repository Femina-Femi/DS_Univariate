```python
#probaility density fucntion
def probdensity(df,start,end):
    import matplotlib.pyplot as plt
    import seaborn as sns
    from scipy.stats import norm
    #plot the given  data
    sns.distplot(df,kde=True,kde_kws = {'color':'blue'},color='green')
    plt.axvline(start,color='Red')
    plt.axvline(end,color='Red')
    #find means and std div for the data
    mean_val=df.mean()
    std_val=df.std()
    print('mean:',mean_val,'std div: ' ,std_val)
    dist = norm(mean_val,std_val)
    #requiredvalues = [value for value in range(start,end)]
    #prob = [dist.pdf(value) for value in requiredvalues ]
    prob = [dist.pdf(values) for values in range(start,end)]
    result = sum(prob)
    return result
    
ans =probdensity(df['ssc_p'],60,70)
print(ans)
```

```python
def stdfn(df):
    import seaborn as sns
    import matplotlib.pyplot as plt
    mean_v=df.mean()
    std_v=df.std()
    z_score=[((x-mean_v)/std_v) for x in df]
    sns.distplot(z_score,kde= True)
    total=sum(z_score)/len(z_score)
    return(total)
stdfn(df['ssc_p'])
```
